# Punto 1 - Implementación de librerías y creación de gráficos con matplotlib

```python
# Importar las librerías de Python que pueden ser de utilidad
import math
import matplotlib.pyplot as plt
import numpy as np

# A) y=sin(1/x)

# Definir el dominio en el eje x para 1000 muestras entre -2 y 2
x1 = np.linspace(-2,2,1000)
# Definir la imagen empleando las funciones incorporadas en Numpy
y1 = np.sin(1/x1)

# Procedimiento básico para plotear una función con arreglos Numpy
plt.plot(x1,y1)
plt.title(f"y=sin(1/x)")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# B) y=x^2 si x<2 & 15-x^2 si x>=2
# Definir el dominio en el eje x para 1000 muestras entre -2 y 6
x2 = np.linspace(-2,6,1000)
y2 = []

# Se establece una función que reciba al dominio y retorne según el tramo de la función
def punto1b(x):
  y = None
  if x < 2:
    y = math.pow(x,2)
  else:
    y = 15-math.pow(x,2)
  return y

# Invocar la función para el dominio establecido entre -2 y 6
for valor_x in x2:
  y2.append(punto1b(valor_x))

# Procedimiento básico para plotear una función con arreglos Numpy
plt.plot(x2,y2)
plt.title("Función por tramos")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# C) z = y*cos(x) en cuadrado [-5,5]x[-5,5]
# Se trabaja con matrices en Numpy al tratarse de una superficie 3D
x = np.outer(np.linspace(-5,5,100), np.ones(100))
y = x.copy().T
z = y*np.cos(x)

# Tomado de StackOverflow para el uso de la función plot_surface
fig = plt.figure(figsize =(14, 14))
ax = plt.axes(projection ='3d')
ax.plot_surface(x, y, z)

# Procedimiento para plotear con Axis
ax.set_title("$z=y*cos(x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
plt.show()
```

# Punto 2 - Implementación de librerías, creación de funciones y uso de condicionales

```python
import numpy as np

# Definir la función de multiplicación maricial
def multiplicar_matrices(A, B, c, d):
  size_A = np.shape(A)
  fil_A = size_A[0]
  col_A = size_A[1]

  size_B = np.shape(B)
  fil_B = size_B[0]
  col_B = size_B[1]

  Apow = A
  Bpow = B
  # Condiciones para manejo del error y la excepción
  if not (type(c) == int) and (type(d) == int):
    return "Los valores c y d no son enteros"
  if fil_A != col_A or fil_B != col_B:
    return "Las matrices A y B deben ser cuadradas para poder multiplicar."
  if col_A != fil_B:
    return "Las dimensiones no permiten la multiplicación de matrices"
  for n in range(c-1):
    Apow = np.matmul(Apow,A)
  for n in range(d-1):
    Bpow = np.matmul(Bpow,B)

  resultado = np.matmul(Apow,Bpow)
  return resultado

# Ejemplo de uso
A = np.array([[0.1, 2],[2, 0.1]])
B = np.array([[1, 2, 3],[4, 5, 6]])
C = np.array([[5/3, 2/3],[5/3, 2/3]])
D = np.array([[5, 4, 3],[6, 2, 9],[1, 7, 8]])


resultado1 = multiplicar_matrices(A, B, 1, 1)
if isinstance(resultado1, str):
    print(resultado1)
else:
    print("Resultado de la multiplicación:")
    for fila in resultado1:
        print(fila)

print("\n")

D_t = np.transpose(D)
resultado2 = multiplicar_matrices(D_t, C, 1, 1)
if isinstance(resultado2, str):
    print(resultado2)
else:
    print("Resultado de la multiplicación:")
    for fila in resultado2:
        print(fila)

print("\n")

resultado3 = multiplicar_matrices(A, C, 2, 6)
if isinstance(resultado3, str):
    print(resultado3)
else:
    print("Resultado de la multiplicación:")
    for fila in resultado3:
        print(fila)
```

# Punto 3 - Implementación de librerías, creación de funciones y recorridos


```python
# Definir la función para determinar si es capicua o no
def es_capicua(x: int) -> str:
  """
  Retorna si un número es capicua (se lee igual de izquierda a derecha y de derecha a izquierda) o no.
  """
  # Emplea listas nativas de Python
  izq_a_der = []
  der_a_izq = []
  intermedio = None
  # Extrae los dígitos del número con el módulo
  while x > 0:
    intermedio = x % 10
    x //= 10
    der_a_izq.append(intermedio)
  # Reversa la lista para hacer la comparación
  izq_a_der = der_a_izq.copy()
  izq_a_der.reverse()

  respuesta = ""
  if izq_a_der == der_a_izq:
    respuesta = "El número SÍ es capicúa."
  else:
    respuesta = "El número NO es capicúa."
  return respuesta

# Banco de pruebas
print("1111: ", es_capicua(1111))
print("121212: ", es_capicua(121212))
print("32123: ", es_capicua(32123))
print("123456: ", es_capicua(123456))
```
